```python
import pandas as pd
import numpy as np
```

```python
titles = pd.read_csv('datasets/title.basics.tsv', sep='\t', na_values='\\N')
```

```python
ratings = pd.read_csv('datasets/title.ratings.tsv', sep='\t', na_values='\\N')
```

```python
crew = pd.read_csv('datasets/title.crew.tsv', sep='\t', na_values='\\N')
```

```python
cast = pd.read_csv('datasets/title.principals.tsv', sep='\t', na_values='\\N')
```

```python
akas = pd.read_csv('datasets/title.akas.tsv', sep='\t', na_values='\\N')
```

```python
#filter only movies
df = titles[(titles['titleType'] == 'movie')]
```

```python
df
```

```python
df.set_index('tconst', inplace=True)
```

```python
ratings.set_index('tconst', inplace=True)
```

```python
ratings
```

```python
#merge df with ratings - left outer join
df = df.merge(ratings, left_on='tconst', right_on='tconst', how='left')
```

```python
crew.set_index('tconst', inplace=True)
```

```python
#merge df with crews - left outer join
df = df.merge(crew, left_on='tconst', right_on='tconst', how='left')
```

```python
df
```

```python
#drop titleType and endYear columns as they are redundant
df.drop(columns=['titleType', 'endYear'], inplace=True)
```

```python
#filter out adult movies
df = df[df['isAdult'] == 0].copy()
```

```python
#drop isAdult column
df.drop(columns=['isAdult'], inplace=True)
```
